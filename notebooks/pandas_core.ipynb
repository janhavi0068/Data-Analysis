# **Rolling Window**

```python
sdf5 =df5.sort_values(by="shipping_limit_date")[["price","freight_value"]]
s=sdf5.rolling(window=4, center=True).sum()
s[:7]
```

```python
s=sdf5.rolling(window=4).sum()
s[:7]
```

```python
s=sdf5.rolling(window=4, closed = "left").sum()
s[:7]
```

In this, the rolling function groups the entire data window with specifies value and applies an aggregation function (here sum).

Calculates sum of 0 with 3 NaN values and provides result as NaN.
2 NaN for 1, 1 NaN with 2nd index. Then calculates [0,1,2,3] sum for 3rd index and [1,2,3,4] for 4th index.

The centre attribute pulls the sum function from 3rd index to 1st (to the centre of window).
Default if False.


The closed attribute considers-
includes the most left data in the window (eg. 4th data in the window of 4 for [_,_,_,1] is NaN is this case).
includes the most right data in the window (eg. In [1,2,3,4], 4th will be considered)
includes both boundaries for "both"
excludes both the boundary data and considers all the data between them for "neither"
